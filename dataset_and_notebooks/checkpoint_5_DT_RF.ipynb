```python
df = pd.read_csv('ecg_df.csv')
for el in (list(df.columns)):
    print(el)
```

```python
import ast

d_of_diseases = {}

for index, row in df.iterrows():
    current_row = ast.literal_eval(row.disease_name)
    for i in range(len(current_row)):
        if current_row[i] == 1:
            if i not in d_of_diseases:
                d_of_diseases[i] = 0
            d_of_diseases[i] += 1
d_of_diseases = {k: v for k, v in sorted(d_of_diseases.items(), key=lambda item: item[1], reverse=True)}
d_of_diseases
```

```python
list_of_diseases = []
for key in d_of_diseases:
    if d_of_diseases[key] > 400:
        list_of_diseases.append(key)
print(list_of_diseases)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, f1_score
count = 0
for disease_id in list_of_diseases:
    count += 1
    print(count,"/" , len(list_of_diseases))
    disease_info = []
    for index, row in df.iterrows():
        lst_row = ast.literal_eval(row.disease_name)
        if lst_row[disease_id] == 2:

            print(row.id, lst_row)
            disease_info.append(1)
        else:
            disease_info.append(lst_row[disease_id])
    df[f'disease{disease_id}'] = disease_info
```

```python
df=df.drop("id", axis = 1)
```

```python
df['gender'] = df['gender'].apply(lambda x: 1 if x == "Male" else 0)
if "signal" in df.columns:
    df = df.drop(columns = ['signal'])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
import time
import optuna

optuna.logging.set_verbosity(optuna.logging.INFO)

target_columns = []
for el in list_of_diseases:
    target_columns.append(f"disease{el}")
target_columns.append("disease_name")

features = df.drop(columns=target_columns)

def get_top_features(model, feature_names, top_n=20):
    importances = model.feature_importances_
    feat_importance_df = pd.DataFrame({
        'feature': feature_names,
        'importance': importances
    })
    top_features = feat_importance_df.sort_values(by='importance', ascending=False).head(top_n)
    return top_features

all_top_features = []
count = 0
for target in target_columns:
    if target == "disease_name":
        continue
    count += 1
    time0 = time.time()
    print(f'Обучение моделей для: {target}    {count}/{len(target_columns)-1}')

    X_train, X_test, y_train, y_test = train_test_split(features, df[target], test_size=0.2, random_state=42)

    rf = RandomForestClassifier(random_state=42)
    dt = DecisionTreeClassifier(random_state=42)

    rf.fit(X_train, y_train)
    dt.fit(X_train, y_train)

    top_rf = get_top_features(rf, X_train.columns, top_n=20)
    top_dt = get_top_features(dt, X_train.columns, top_n=20)

    all_top_features.append(top_rf)
    all_top_features.append(top_dt)

    print('-------------------------------------------')

df_all_top = pd.concat(all_top_features, ignore_index=True)

df_grouped = df_all_top.groupby('feature', as_index=False).mean()

df_grouped = df_grouped.sort_values(by='importance', ascending=False)
final_top20_features = df_grouped.head(20)['feature'].tolist()
final_top20_features
```

```python
optuna.logging.set_verbosity(optuna.logging.CRITICAL)

models_rf = {}
models_dt = {}

def get_best_models(x, y):
    X_train_tmp, X_valid_tmp, y_train_tmp, y_valid_tmp = train_test_split(x, y, test_size=0.2, random_state=42)

    def objective_rf(trial):
        n_estimators = trial.suggest_int('n_estimators', 50, 300)
        max_depth = trial.suggest_int('max_depth', 3, 20)
        min_samples_split = trial.suggest_int('min_samples_split', 2, 10)

        rf = RandomForestClassifier(
            random_state=42,
            n_estimators=n_estimators,
            max_depth=max_depth,
            min_samples_split=min_samples_split
        )
        rf.fit(X_train_tmp, y_train_tmp)
        preds = rf.predict(X_valid_tmp)
        score = f1_score(y_valid_tmp, preds, average='weighted')
        return score

    study_rf = optuna.create_study(direction='maximize')
    study_rf.optimize(objective_rf, n_trials=20)

    def objective_dt(trial):
        max_depth = trial.suggest_int('max_depth', 3, 20)
        min_samples_split = trial.suggest_int('min_samples_split', 2, 10)

        dt = DecisionTreeClassifier(
            random_state=42,
            max_depth=max_depth,
            min_samples_split=min_samples_split
        )
        dt.fit(X_train_tmp, y_train_tmp)
        preds = dt.predict(X_valid_tmp)
        score = f1_score(y_valid_tmp, preds, average='binary')
        return score

    study_dt = optuna.create_study(direction='maximize')
    study_dt.optimize(objective_dt, n_trials=20)
    return [study_rf.best_trial.params, study_dt.best_trial.params]

for target in target_columns:
    time0 = time.time()
    if target == "disease_name":
        continue
    print(f'Обучение моделей для: {target}')

    X_train, X_test, y_train, y_test = train_test_split(features[final_top20_features], df[target], test_size=0.2, random_state=42)

    rf_params, df_params = get_best_models(X_train, y_train)

    rf = RandomForestClassifier(**rf_params)
    dt = DecisionTreeClassifier(**df_params)

    rf.fit(X_train, y_train)
    dt.fit(X_train, y_train)

    models_rf[target] = rf
    models_dt[target] = dt

    y_pred_rf = rf.predict(X_test)
    y_pred_dt = dt.predict(X_test)

    f1_rf = f1_score(y_test, y_pred_rf, average='weighted')
    f1_dt = f1_score(y_test, y_pred_dt, average='weighted')

    print('RandomForestClassifier:')
    print(f"F1-score: {f1_rf}")
    print(f"Best params: {rf_params}")
    print('\nDecisionTreeClassifier:')
    print(f"F1-score: {f1_dt}")
    print(f"Best params: {df_params}")
    print(f"{time.time() - time0} секунд")
    print('-------------------------------------------')
```
